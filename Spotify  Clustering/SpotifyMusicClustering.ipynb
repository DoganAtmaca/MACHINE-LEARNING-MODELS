BU PROJEDE MÜZİK ÖZELLİKLERİNE GÖRE MÜZİKLERİ KÜMELERE AYIRACAĞIZ.

```python
import pandas as pd
import warnings
```

```python
#UYARILARI GİZLİYORUZ.
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv('Spotify-2000.csv')
```

```python
data.head()
```

```python
data = data.drop("Index", axis=1)
```

```python
data.info()
```

```python
#SES ÖZELLİKLERİNE GÖRE KÜMELEME YAPACAĞIMIZ İÇİN AŞAĞIDAKİ SÜTUNLARI ÇIKARIYORUZ. BUNLAR SES ÖZELLİKLERİ İLE ALAKALI DEĞİL.
data2 = data.drop(['Title','Artist','Top Genre','Length (Duration)','Year','Popularity','Energy','Danceability'],axis=1)
```

```python
#SAYISAL VERİNİN KORELASYONUNU İNCELİYORUZ.
import seaborn as sns
import matplotlib.pyplot as plt

corr = data2.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
```

```python
#GEREKLİ KÜTÜPHANELERİ İMPORT EDİYORUZ. VERİYİ SCALE ETMEK VE KAÇ TANE KÜME YAPACAĞIMIZI BULMAK İÇİN.
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from yellowbrick.cluster import KElbowVisualizer
```

```python
scaler = StandardScaler()
data_scale = scaler.fit_transform(data2)
```

```python
#KAÇ KÜME YAPILMASININ DAHA UYGUN OLACAĞINI GÖRSELLEŞTİRELİM
km = KMeans()
vis = KElbowVisualizer(km,k = (2,15)) # BURADA 2 DEN 15 E KADAR DENEME YAPILACAK. 
vis.fit(data_scale)
vis.show()
```

```python
#6 TANE KÜME OLMASININ EN UYGUN OLACAĞINI GÖRÜYORUZ. BUNUN ÜZERİNE ARTIK KÜMELEME İŞLEMİNE GEÇEBİLİRİZ.
model = KMeans(6)
model = model.fit(data_scale)
tahmin = model.predict(data_scale)
```

```python
#OLUŞTURDUĞUMUZ KÜMELERİ YENİ BİR SÜTUNDA GÖSTERİYORUZ.
data['Cluster'] = tahmin
```

```python
data.head()
```

```python
#OLUŞTURDUĞUMUZ KÜMELERİ GÖRSELLEŞTİRİYORUZ.
import plotly.graph_objects as go

PLOT = go.Figure()

for i in data["Cluster"].unique():
    PLOT.add_trace(go.Scatter3d(
        x=data[data["Cluster"] == i]['Beats Per Minute (BPM)'],
        y=data[data["Cluster"] == i]['Energy'],
        z=data[data["Cluster"] == i]['Danceability'],
        mode='markers',
        marker=dict(size=6, line=dict(width=1)),
        name=f"Cluster {i}"
    ))

PLOT.update_traces(
    hovertemplate='Beats Per Minute (BPM): %{x}<br>Energy: %{y}<br>Danceability: %{z}'
)

PLOT.update_layout(
    width=800,
    height=800,
    autosize=True,
    showlegend=True,
    scene=dict(
        xaxis=dict(
            title=dict(text='Beats Per Minute (BPM)', font=dict(color='black'))
        ),
        yaxis=dict(
            title=dict(text='Energy', font=dict(color='black'))
        ),
        zaxis=dict(
            title=dict(text='Danceability', font=dict(color='black'))
        )
    ),
    font=dict(family="Gilroy", color='black', size=12)
)

PLOT.show()
```

```python
#EN SON SILHOUETTE SCORE HESAPLAYIP MODELİN BAŞARISINI ÖLÇÜYORUZ.
from sklearn.metrics import silhouette_score

score = silhouette_score(data_scale, model.labels_)
print("Silhouette Score:", score)
```
